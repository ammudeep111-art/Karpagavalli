```python
import pandas as pd
import numpy as np
import warnings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score,accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('Data/StudentPerformanceFactors.csv')
```

```python
data
```

```python
data.shape
```

```python
data.drop(['Parental_Involvement','Access_to_Resources','Internet_Access','Family_Income',
          'Teacher_Quality','School_Type','Peer_Influence','Learning_Disabilities'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['Extracurricular_Activities']=data['Extracurricular_Activities'].map({'Yes':1,'No':0})
data['Motivation_Level']=data['Motivation_Level'].map({'Low':0,'Medium':1,'High':2})
data['Gender']=data['Gender'].map({'Male':1,'Female':0})
```

```python
data.drop(['Parental_Education_Level','Distance_from_Home'],axis=1,inplace=True)
```

```python
X = data[['Hours_Studied','Attendance','Extracurricular_Activities','Sleep_Hours','Motivation_Level',
         'Tutoring_Sessions','Physical_Activity','Previous_Scores','Gender',]]
y = data[['Exam_Score']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```
